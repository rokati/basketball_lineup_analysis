```python
import pandas as pd
```

# Step 1: Load the CSV

```python
df = pd.read_csv('..\\updated_lineup_vs_lineup\\lineup_vs_lineup_with_team_id.csv')
```

```python
df = df.fillna(0)
```

```python
df
```

# Step 2: Define the formula for lineup_score

```python
# Time conversion helper (from MM:SS to total seconds)
def time_to_seconds(t):
    minutes, seconds = map(int, str(t).split(":"))
    return minutes * 60 + seconds

# Convert start_time and end_time to seconds
df['start_sec'] = df['start_time'].apply(time_to_seconds)
df['end_sec'] = df['end_time'].apply(time_to_seconds)
df['duration'] = df['start_sec'] - df['end_sec']  # time moving backward (start_time > end_time)
```

```python
df.head()
```

```python
df['home_score'] = (
    1.0 * df['home_points'] +
    0.7 * df['home_assists'] +
    1.0 * df['home_3pt_made'] +
    1.0 * df['home_2pt_made'] +
    0.5 * df['home_ft_made'] +
    0.7 * df['home_steals'] +
    0.5 * df['home_blocks'] +
    0.3 * df['home_def_rebounds'] +
    0.3 * df['home_off_rebounds'] +
    (-1.0) * df['home_turnovers'] +
    (-0.3) * df['home_fouls']
)

df['away_score'] = (
    1.0 * df['away_points'] +
    0.7 * df['away_assists'] +
    1.0 * df['away_3pt_made'] +
    1.0 * df['away_2pt_made'] +
    0.5 * df['away_ft_made'] +
    0.7 * df['away_steals'] +
    0.5 * df['away_blocks'] +
    0.3 * df['away_def_rebounds'] +
    0.3 * df['away_off_rebounds'] +
    (-1.0) * df['away_turnovers'] +
    (-0.3) * df['away_fouls']
)
```

```python
df['normalized_home_score'] = df['home_score'] / df['duration']
df['normalized_away_score'] = df['away_score'] / df['duration']
df['net_score'] = df['normalized_home_score'] - df['normalized_away_score']
```

```python
df
```

# Step 3: Save to a new CSV

```python
df.to_csv('lineup_score_with_team_id.csv', index=False)

print("✅ New file saved as 'lineup_score_with_team_id.csv'")
```
